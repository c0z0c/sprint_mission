### üá∂ ÏÑ†ÌòïÎåÄÏàòÌïôÏù¥ÎûÄ Ïñ¥Îñ§ ÌïôÎ¨∏Ïù¥Î©∞, Ïôú Î®∏Ïã†Îü¨ÎãùÏóê ÌïÑÏöîÌïúÏßÄ ÏÑ§Î™ÖÌï¥Î≥¥ÏÑ∏Ïöî.
#### 1. ÏÑ†ÌòïÎåÄÏàòÌïôÏù¥ÎûÄ?
- Î≤°ÌÑ∞, ÌñâÎ†¨, ÏÑ†ÌòïÎ≥ÄÌôò, Í≥†Ïú†Í∞í Î∂ÑÌï¥, ÎÇ¥Ï†ÅÎì±Ïùò Í∞úÎÖêÏúºÎ°ú Íµ¨ÏÑ±ÎêòÏñ¥ ÏûàÎã§.
- Ï∞∏Í≥† : Ïù∏Í≥µÏßÄÏùëÏùÑ ÏúÑÌïú ÏÑ†ÌòïÎåÄÏàòÌïô Í∏∞Ï¥à (https://wikidocs.net/214391)


- Î∞±ÌÑ∞ / ÌñâÎ†¨ / ÎÇ¥Ï†Å
    - np.array([ ... ]) :
        - Î¶¨Ïä§Ìä∏Î•º ÎÑòÌååÏù¥ Î∞∞Ïó¥(Î≤°ÌÑ∞)Î°ú Î≥ÄÌôòÌï©ÎãàÎã§.
    - v1 + v2 :
        - Îëê Î≤°ÌÑ∞Ïùò Í∞Å ÏõêÏÜåÎ•º ÎçîÌï©ÎãàÎã§(Î≤°ÌÑ∞ ÎçßÏÖà).
    - np.dot(v1, v2) :
        - Îëê Î≤°ÌÑ∞Ïùò ÎÇ¥Ï†Å(Ïä§ÏπºÎùº Í≥±)ÏùÑ Í≥ÑÏÇ∞Ìï©ÎãàÎã§.
    - plt.quiver(...) :
        - 2Ï∞®Ïõê ÌèâÎ©¥Ïóê Î≤°ÌÑ∞(ÌôîÏÇ¥Ìëú)Î•º Í∑∏Î¶ΩÎãàÎã§.
    - plt.xlim(), plt.ylim() :
        - Í∑∏ÎûòÌîÑÏùò xÏ∂ï, yÏ∂ï Î≤îÏúÑÎ•º ÏÑ§Ï†ïÌï©ÎãàÎã§.
    - plt.grid() :
        - Í∑∏ÎûòÌîÑÏóê Í≤©Ïûê(Í∑∏Î¶¨Îìú)Î•º ÌëúÏãúÌï©ÎãàÎã§.
    - plt.legend() :
        - Í∑∏ÎûòÌîÑÏóê Î≤îÎ°Ä(legend)Î•º Ï∂îÍ∞ÄÌï©ÎãàÎã§.
    - plt.title() :
        - Í∑∏ÎûòÌîÑÏùò Ï†úÎ™©ÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
    - plt.show() :
        - Í∑∏ÎûòÌîÑÎ•º ÌôîÎ©¥Ïóê Ï∂úÎ†•Ìï©ÎãàÎã§.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


v1 = np.array([2, 3])
v2 = np.array([1, 1])

v_sum = v1 + v2

# ÎÇ¥Ï†Å
dot = v1 @ v2

# Ïä§ÏπºÎùº Í≥±
v12 = 2 * v1

# ÌñâÎ†¨ Ï†ÑÏπò
v1_T = v1.T

print("v1 =", v1)
print("v2 =", v2)
print("v1 + v2 =", v_sum)
print("ÎÇ¥Ï†Å v1 @ v2 =", dot)
print("Ïä§ÏπºÎùº 2 * v1 =", v12)
print("Ï†ÑÏπò v1_T =", v1_T)

origin = [0, 0]
plt.figure(figsize=(4, 3))
plt.quiver(*origin, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(*origin, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')
plt.quiver(*origin, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='g', label='v1+v2')

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid()
plt.legend()
plt.title('Î≤°ÌÑ∞Ïùò ÏãúÍ∞ÅÌôî')
plt.show()
```

```python
import numpy as np

# ÌñâÎ†¨ ÏÉùÏÑ±
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# ÌñâÎ†¨ ÎçßÏÖà
C = A + B
# ÌñâÎ†¨ Í≥±ÏÖà
D = A @ B
# Ïä§ÏπºÎùº Í≥±
E = 2 * A
# ÌñâÎ†¨ Ï†ÑÏπò
A_T = A.T

print('A =\n', A)
print('B =\n', B)
print('A + B =\n', C)
print('A * B (ÌñâÎ†¨Í≥±) =\n', D)
print('2 * A Ïä§ÏπºÎùº Í≥± =\n', E)
print('AÏùò Ï†ÑÏπò =\n', A_T)
```

- Í≥†Ïú†Í∞í Î∂ÑÌï¥
    - numpy.linalg.eig(a)
        - Ï†ïÎ∞©ÌñâÎ†¨(Ï†ïÏÇ¨Í∞ÅÌòï ÌñâÎ†¨)Ïùò Í≥†Ïú≥Í∞í(eigenvalue)Í≥º Í≥†Ïú†Î≤°ÌÑ∞(eigenvector)Î•º Í≥ÑÏÇ∞Ìï¥Ï£ºÎäî Ìï®ÏàòÏûÖÎãàÎã§.
        - a : Ï†ïÎ∞©ÌñâÎ†¨ (shape (n,n))
        - Î∞òÌôòÍ∞í (w, v)
            - w : Í≥†Ïú†Í∞í Î∞∞Ïó¥, shape (n,)
            - v : Í≥†Ïú†Î≤°ÌÑ∞Îì§, shape (n,n) ‚Üí v[:, i] Í∞Ä w[i]Ïóê ÎåÄÏùëÌïòÎäî Í≥†Ïú†Î≤°ÌÑ∞


```python
import numpy as np

# 2x2 ÎåÄÍ∞ÅÌñâÎ†¨
A = np.array([[2, 0],
              [0, 3]])

# Í≥†Ïú†Í∞í, Í≥†Ïú†Î≤°ÌÑ∞ Í≥ÑÏÇ∞
A1, V = np.linalg.eig(A)

print("ÌñâÎ†¨ A:\n", A)
print("Í≥†Ïú†Í∞í:", A1)
print("Í≥†Ïú†Î≤°ÌÑ∞:\n", V)
```

### Í≤∞Ï∏°Í∞í
- ÏÇ≠Ï†ú ÌòπÏùÄ ÏàòÏ†ï

```python
# Í≤∞Ï∏°Í∞í ÌôïÏù∏
df = titanic_org.copy()

# ÌïúÍ∏Ä Ïª¨ÎüºÎ™ÖÏúºÎ°ú Î≥ÄÍ≤Ω
kr_labels = df.get_head_att()
df.columns = [kr_labels.get(col, col) for col in df.columns]

# Í≤∞Ï∏°Í∞í ÌûàÌä∏Îßµ
plt.figure(figsize=(8,3))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False)
plt.xticks(rotation=-80)
plt.title('Í≤∞Ï∏°Í∞í Î∂ÑÌè¨')
plt.show()

# Í≤∞Ï∏°Í∞í Í∞úÏàò Ï∂úÎ†•
print("Í≤∞Ï∏°Í∞í Í∞úÏàò:")
print(df.isnull().sum().sort_values(ascending=False).head(5))
```

```python
# ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ ÏõêÎ≥µ
df = titanic_org.copy()

# Î∂àÌïÑÏöîÌïú Ìï≠Î™© ÏÇ≠Ï†ú
df.drop(columns=['deck'], inplace=True)

print("=== Ï§ëÎ≥µ Îç∞Ïù¥ÌÑ∞ Ï≤òÎ¶¨ ===")
print("ÏõêÎ≥∏ Ï¥ù Ìñâ Í∞úÏàò :", len(df))

# Ï§ëÎ≥µ Ìñâ Í∞úÏàò ÌôïÏù∏
duplicate_count = df.duplicated().sum()
print("Ï§ëÎ≥µ Ìñâ Í∞úÏàò:", duplicate_count)

if duplicate_count > 0:
    # Ï§ëÎ≥µ Ìñâ ÎØ∏Î¶¨ Î≥¥Í∏∞
    print("-" * 80)
    duplicate_rows = df[df.duplicated()]
    duplicate_rows.head_att(5)

    # Ï§ëÎ≥µ Ìñâ ÏÇ≠Ï†ú (Ï≤´ Î≤àÏß∏Îßå ÎÇ®Í∏∞Í≥† ÏÇ≠Ï†ú)
    df = df.drop_duplicates()
    print("-" * 80)
    print("ÏÇ≠Ï†ú ÌõÑ Ï¥ù Ìñâ Í∞úÏàò :", len(df))
else:
    print("Ï§ëÎ≥µ ÌñâÏù¥ ÏóÜÏäµÎãàÎã§.")

print("\n=== ÎÇòÏù¥ Í≤∞Ï∏°Í∞í Ï≤òÎ¶¨ ===")

# ÏÑ±Ïù∏/ÎπÑÏÑ±Ïù∏ Í∑∏Î£πÎ≥Ñ ÌèâÍ∑† ÎÇòÏù¥ Í≥ÑÏÇ∞
adult_mean = df.loc[df['adult_male'] == True, 'age'].mean()
not_adult_mean = df.loc[df['adult_male'] == False, 'age'].mean()

print(f"ÏÑ±Ïù∏ Í∑∏Î£π ÌèâÍ∑† ÎÇòÏù¥: {adult_mean:.2f}")
print(f"ÎπÑÏÑ±Ïù∏ Í∑∏Î£π ÌèâÍ∑† ÎÇòÏù¥: {not_adult_mean:.2f}")

# Í≤∞Ï∏°Í∞íÏùÑ Í∑∏Î£πÎ≥Ñ ÌèâÍ∑†ÏúºÎ°ú ÎåÄÏ≤¥
df.loc[(df['age'].isnull()) & (df['adult_male'] == True), 'age'] = adult_mean
df.loc[(df['age'].isnull()) & (df['adult_male'] == False), 'age'] = not_adult_mean

print("\nÍ≤∞Ï∏°Í∞í Ï≤òÎ¶¨ ÌõÑ:")
print(df.isnull().sum().sort_values(ascending=False).head(5))

# ÌÖåÏä§Ìä∏Î•º ÏúÑÌïòÏó¨ Î≥µÏÇ¨
titanic_eda_drop_org = df.copy()

print("\n=== Ï≤òÎ¶¨Îêú Îç∞Ïù¥ÌÑ∞ ÎØ∏Î¶¨Î≥¥Í∏∞ ===")

df.head_att(5,'html')

```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = titanic_eda_drop_org.copy()

# Ïà´ÏûêÌòï Ïª¨ÎüºÎßå ÏÑ†ÌÉù
numeric_cols = df.select_dtypes(include=['number']).columns
print(f"Ïà´ÏûêÌòï Ïª¨ÎüºÎì§: {list(numeric_cols)}")

# 4Í∞úÏî© ÎÇòÎàÑÍ∏∞
split_cols = np.array_split(numeric_cols, (len(numeric_cols) + 3) // 4)

for i, cols in enumerate(split_cols):
    plt.figure(figsize=(4 * len(cols), 5))

    for j, col in enumerate(cols):
        plt.subplot(1, len(cols), j + 1)

        # Í≤∞Ï∏°Í∞í Ï†úÍ±∞ÌïòÍ≥† Î∞ïÏä§ÌîåÎ°Ø ÏÉùÏÑ±
        data_clean = df[col].dropna()
        if len(data_clean) > 0:
            plt.boxplot(data_clean)
            plt.title(f'{col} Î∞ïÏä§ÌîåÎ°Ø')  # ÌïúÍ∏Ä Ï†úÎ™©
            plt.ylabel('Í∞í')
        else:
            plt.text(0.5, 0.5, 'Îç∞Ïù¥ÌÑ∞ ÏóÜÏùå', ha='center', va='center')
            plt.title(f'{col} (Îç∞Ïù¥ÌÑ∞ ÏóÜÏùå)')

    plt.tight_layout()
    plt.suptitle(f"ÌÉÄÏù¥ÌÉÄÎãâ Îç∞Ïù¥ÌÑ∞ Î∞ïÏä§ÌîåÎ°Ø (Í∑∏Î£π {i+1})", y=1.02)
    plt.show()

print("‚úÖ Î∞ïÏä§ÌîåÎ°Ø ÏÉùÏÑ± ÏôÑÎ£å")
```
