### 🇶 선형대수학이란 어떤 학문이며, 왜 머신러닝에 필요한지 설명해보세요.
#### 1. 선형대수학이란?
- 벡터, 행렬, 선형변환, 고유값 분해, 내적등의 개념으로 구성되어 있다.
- 참고 : 인공지응을 위한 선형대수학 기초 (https://wikidocs.net/214391)


- 백터 / 행렬 / 내적
    - np.array([ ... ]) :
        - 리스트를 넘파이 배열(벡터)로 변환합니다.
    - v1 + v2 :
        - 두 벡터의 각 원소를 더합니다(벡터 덧셈).
    - np.dot(v1, v2) :
        - 두 벡터의 내적(스칼라 곱)을 계산합니다.
    - plt.quiver(...) :
        - 2차원 평면에 벡터(화살표)를 그립니다.
    - plt.xlim(), plt.ylim() :
        - 그래프의 x축, y축 범위를 설정합니다.
    - plt.grid() :
        - 그래프에 격자(그리드)를 표시합니다.
    - plt.legend() :
        - 그래프에 범례(legend)를 추가합니다.
    - plt.title() :
        - 그래프의 제목을 설정합니다.
    - plt.show() :
        - 그래프를 화면에 출력합니다.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


v1 = np.array([2, 3])
v2 = np.array([1, 1])

v_sum = v1 + v2

# 내적
dot = v1 @ v2

# 스칼라 곱
v12 = 2 * v1

# 행렬 전치
v1_T = v1.T

print("v1 =", v1)
print("v2 =", v2)
print("v1 + v2 =", v_sum)
print("내적 v1 @ v2 =", dot)
print("스칼라 2 * v1 =", v12)
print("전치 v1_T =", v1_T)

origin = [0, 0]
plt.figure(figsize=(4, 3))
plt.quiver(*origin, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(*origin, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')
plt.quiver(*origin, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='g', label='v1+v2')

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid()
plt.legend()
plt.title('벡터의 시각화')
plt.show()
```

- 행렬


```python
import numpy as np

# 행렬 생성
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# 행렬 덧셈
C = A + B
# 행렬 곱셈
D = A @ B
# 스칼라 곱
E = 2 * A
# 행렬 전치
A_T = A.T

print('A =\n', A)
print('B =\n', B)
print('A + B =\n', C)
print('A * B (행렬곱) =\n', D)
print('2 * A 스칼라 곱 =\n', E)
print('A의 전치 =\n', A_T)
```

```python
import numpy as np
# import matplotlib.pyplot as plt

print('행렬을 이용한 "크기 2배 늘리기" (스케일 변환)')
print('2차원 벡터 [x, y]를 x, y 모두 2배로 늘리는 변환입니다.')
print()

v = np.array([1, 2])  # 원래 벡터
S = np.array([[2, 0], [0, 2]])  # 스케일(2배) 행렬

v_scaled = S @ v  # 변환 적용

print("원래 벡터:", v)
print("스케일 행렬 S:\n", S)
print("2배로 늘린 벡터:", v_scaled)

# origin = [0, 0]
# plt.figure(figsize=(4, 4))
# plt.quiver(*origin, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='원래 벡터')
# plt.xlim(-1, 5)
# plt.ylim(-1, 5)
# plt.grid()
# plt.legend()
# plt.title("원래 백터")
# plt.show()

# origin = [0, 0]
# plt.figure(figsize=(4, 4))
# plt.quiver(*origin, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='원래 벡터')
# plt.quiver(*origin, v_scaled[0], v_scaled[1], angles='xy', scale_units='xy', scale=1, color='g', label='2배 벡터')
# plt.xlim(-1, 5)
# plt.ylim(-1, 5)
# plt.grid()
# plt.legend()
# plt.title('행렬로 2배 늘리기(스케일 변환)')
# plt.show()

origin = [0, 0]
plt.figure(figsize=(8, 4))

# 첫 번째 subplot: 원래 벡터
plt.subplot(1, 2, 1)
plt.quiver(*origin, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='원래 벡터')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid()
plt.legend()
plt.title("원래 벡터")

# 두 번째 subplot: 원래 벡터 + 2배 벡터
plt.subplot(1, 2, 2)
plt.quiver(*origin, v_scaled[0], v_scaled[1], angles='xy', scale_units='xy', scale=1, color='g', label='2배 벡터')
plt.quiver(*origin, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='원래 벡터')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid()
plt.legend()
plt.title('행렬로 2배 늘리기(스케일 변환)')

plt.tight_layout()
plt.show()

```

- 고유값 분해
    - numpy.linalg.eig(a)
        - 정방행렬(정사각형 행렬)의 고윳값(eigenvalue)과 고유벡터(eigenvector)를 계산해주는 함수입니다.
        - a : 정방행렬 (shape (n,n))
        - 반환값 (w, v)
            - w : 고유값 배열, shape (n,)
            - v : 고유벡터들, shape (n,n) → v[:, i] 가 w[i]에 대응하는 고유벡터


```python
import numpy as np

# 2x2 대각행렬
A = np.array([[2, 0],
              [0, 3]])

# 고유값, 고유벡터 계산
A1, V = np.linalg.eig(A)

print("행렬 A:\n", A)
print("고유값:", A1)
print("고유벡터:\n", V)
```

### 결측값
- 삭제 혹은 수정
- 불필요한 행 삭제
- 평균, 중앙값 등으로 대체

```python
# 결측값 확인
df = titanic_org.copy()

# 한글 컬럼명으로 변경
kr_labels = df.get_head_att()
df.columns = [kr_labels.get(col, col) for col in df.columns]

# 결측값 히트맵
plt.figure(figsize=(8,3))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False)
plt.xticks(rotation=-80)
plt.title('결측값 분포')
plt.show()

# 결측값 개수 출력
print("결측값 개수:")
print(df.isnull().sum().sort_values(ascending=False).head(5))
```

```python
# 테스트 데이터 원복
df = titanic_org.copy()

# 불필요한 항목 삭제
df.drop(columns=['deck'], inplace=True)

print("=== 중복 데이터 처리 ===")
print("원본 총 행 개수 :", len(df))

# 중복 행 개수 확인
duplicate_count = df.duplicated().sum()
print("중복 행 개수:", duplicate_count)

if duplicate_count > 0:
    # 중복 행 미리 보기
    print("-" * 80)
    duplicate_rows = df[df.duplicated()]
    duplicate_rows.head_att(5)

    # 중복 행 삭제 (첫 번째만 남기고 삭제)
    df = df.drop_duplicates()
    print("-" * 80)
    print("삭제 후 총 행 개수 :", len(df))
else:
    print("중복 행이 없습니다.")

print("\n=== 나이 결측값 처리 ===")

# 성인/비성인 그룹별 평균 나이 계산
adult_mean = df.loc[df['adult_male'] == True, 'age'].mean()
not_adult_mean = df.loc[df['adult_male'] == False, 'age'].mean()

print(f"성인 그룹 평균 나이: {adult_mean:.2f}")
print(f"비성인 그룹 평균 나이: {not_adult_mean:.2f}")

# 결측값을 그룹별 평균으로 대체
df.loc[(df['age'].isnull()) & (df['adult_male'] == True), 'age'] = adult_mean
df.loc[(df['age'].isnull()) & (df['adult_male'] == False), 'age'] = not_adult_mean

print("\n결측값 처리 후:")
print(df.isnull().sum().sort_values(ascending=False).head(5))

# 테스트를 위하여 복사
titanic_eda_drop_org = df.copy()

print("\n=== 처리된 데이터 미리보기 ===")

df.head_att(5,'html')

```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = titanic_eda_drop_org.copy()

# 숫자형 컬럼만 선택
numeric_cols = df.select_dtypes(include=['number']).columns
print(f"숫자형 컬럼들: {list(numeric_cols)}")

# 4개씩 나누기
split_cols = np.array_split(numeric_cols, (len(numeric_cols) + 3) // 4)

for i, cols in enumerate(split_cols):
    plt.figure(figsize=(4 * len(cols), 5))

    for j, col in enumerate(cols):
        plt.subplot(1, len(cols), j + 1)

        # 결측값 제거하고 박스플롯 생성
        data_clean = df[col].dropna()
        if len(data_clean) > 0:
            plt.boxplot(data_clean)
            plt.title(f'{col} 박스플롯')  # 한글 제목
            plt.ylabel('값')
        else:
            plt.text(0.5, 0.5, '데이터 없음', ha='center', va='center')
            plt.title(f'{col} (데이터 없음)')

    plt.tight_layout()
    plt.suptitle(f"타이타닉 데이터 박스플롯 (그룹 {i+1})", y=1.02)
    plt.show()

print("✅ 박스플롯 생성 완료")
```
